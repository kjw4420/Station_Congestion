```python
import pandas as pd
import numpy as np

data = pd.read_csv('지하철혼잡도_2024.csv') 
df = pd.DataFrame(data)

df.head() #데이터 확인
```

```python
# 출근시간대, 퇴근시간대
columns = [
    '요일구분', '호선', '출발역', '상하구분',
    '7시00분', '7시30분', '8시00분', '8시30분', '9시00분',
    '17시00분', '17시30분', '18시00분', '18시30분', '19시00분'
]

# 데이터프레임에서 필요한 컬럼만 필터링
df = df[columns]

print(df)
```

```python
# 요일 구분 == 평일

filtered_df = df[df['요일구분'] == '평일']

print(filtered_df)
```

```python
# 출근시간 데이터 생성: 출근시간대 혼잡도의 평균
morning = ['7시00분', '7시30분', '8시00분', '8시30분', '9시00분']
df['출근시간_혼잡도'] = df[morning].mean(axis=1)

df.drop(columns=morning, inplace=True)

# # 결과 출력
print(df)
```

```python
# 퇴근시간 데이터 생성: 퇴근시간대 혼잡도의 평균
evening = ['17시00분',  '17시30분',  '18시00분',  '18시30분',  '19시00분']
df['퇴근시간_혼잡도'] = df[evening].mean(axis=1)

df.drop(columns=evening, inplace=True)

# # 결과 출력
print(df)
```

```python
p = df[df['출발역']== '군자']

print(p)
```

```python
# 출근시간 혼잡한 역 상위 5개

top5_morning = df.nlargest(5, '출근시간_혼잡도')

print(top5_morning)
```

```python
# 출근시간 파일 생성 및 저장
top5_morning.drop(columns='퇴근시간_혼잡도', inplace=True)

# '출근시간_혼잡도': 소수점 둘째 자리로 포맷
top5_morning['출근시간_혼잡도'] = top5_morning['출근시간_혼잡도'].map('{:.2f}'.format)
print(top5_morning)

# CSV 파일로 저장
top5_morning.to_csv('top5_morning.csv', index=False)
```

```python
# 퇴근시간 혼잡한 역 TOP 5
top5_evening = df.nlargest(5, '퇴근시간_혼잡도')

print(top5_evening)
```

```python
# 퇴근시간 파일 생성 및 저장
top5_evening.drop(columns='출근시간_혼잡도', inplace=True)

# '퇴근시간_혼잡도': 소수점 둘째 자리로 포맷
top5_evening['퇴근시간_혼잡도'] = top5_evening['퇴근시간_혼잡도'].map('{:.2f}'.format)
print(top5_evening)

# CSV 파일로 저장
top5_evening.to_csv('top5_evening.csv', index=False)
```
